```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

df = sns.load_dataset('titanic')
```

# Univariate Analysis

```python
#Q. How many survived vs how many died
# Countplot , barplot
sns.countplot(x='survived',data=df)
plt.xticks([0,1],['Died','Survived'])
```

*more people died as compare to survived*

Right skewed graph

```python
#age distribution dropped missing value
sns.histplot(df['age'].dropna(),kde=True)
```

```python
# Age distribution
df['age']=df['age'].fillna(df['age'].median())
```

```python
sns.histplot(df['age'],kde=True)
```

```python
sns.histplot(df['fare'].dropna(),kde=True)
```

Observation:

Age: Roughly bell-shaped, peaking around 20-30 years old. Note the small bump near 0-5 (infants).

Fare: Highly Right-Skewed. Most tickets were cheap, but a few were incredibly expensive (Outliers).

#BIVARIATE  ANALYSIS

```python
# Hypothesis: Did gender matter?
# Which gender survived more?
sns.barplot(x='sex',y='survived',data=df)
```

```python
# Hypothesis: Rich people survived more, 1st class, 3rd class
sns.barplot(x='pclass',y='survived',data=df)
```

1 class people more survived

```python
# Hypothesis: Did age matter in survival?
sns.boxplot(x='survived',y='age',data=df)
```

Age did not matter that much

```python
#feature engineering
df['family_size']=df['sibsp']+df['parch']+1
```

```python
#hypothesis matter
sns.barplot(x='family_size', y='survived', data = df)
```
